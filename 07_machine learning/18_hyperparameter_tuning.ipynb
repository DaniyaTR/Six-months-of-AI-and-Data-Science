# hyperparameter tuning

Hyperparameter tuning is the process of selecting the best hyperparameters—model settings not learned from training data (e.g., learning rate, regularization strength, tree depth, number of neighbors)—to maximize model performance on validation data, typically using methods like grid search, random search, or Bayesian optimization.

# Cross-validation

Cross-validation is a resampling method to evaluate a model's generalization by splitting data into multiple train/validation folds (e.g., k-fold), training on subsets and validating on held-out folds, then averaging performance—used to choose models, estimate performance reliably, and reduce overfitting.
