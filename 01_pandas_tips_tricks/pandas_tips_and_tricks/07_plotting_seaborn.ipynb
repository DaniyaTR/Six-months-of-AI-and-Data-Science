add caps to error bars

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('titanic')

sns.set_style('darkgrid')  # other styles are dark, white, ticks
# Scatter plot size of figure, we adjust size of figure on our own choice
plt.figure(figsize=(12,6))  # this 12,8 are widht & height of figure
# change the scatter points color
palette={'First':'black', 'Second':'pink', 'Third':'green'}
sns.boxplot(data=df, x='sex',y='age',hue='class', palette=palette )
# add caps to error bars
plt.errorbar(x=[-0.2,0.8], y=[30,25], yerr=[5,5], fmt='o', color='red', capsize=10)
```

## Violin plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('titanic')

sns.set_style('darkgrid')  # other styles are dark, white, ticks
# Scatter plot size of figure, we adjust size of figure on our own choice
plt.figure(figsize=(12,6))  # this 12,8 are widht & height of figure
# change the scatter points color
palette={'First':'black', 'Second':'pink', 'Third':'green'}
sns.violinplot(data=df, x='sex',y='age',hue='class', palette=palette )
# add caps to error bars
plt.errorbar(x=[-0.2,0.8], y=[30,25], yerr=[5,5], fmt='o', color='red', capsize=10)
```
